```python
def get_price_seg(price):
    if price < 501:
        return 'Low'
    elif price < 1001:
        return 'Medium'
    elif price < 1501:
        return 'High medium'
    elif price >= 1501:
        return 'Premium'
    else:
        return price

#Запустили функцию для добавления нового столбца
```

```python
wine['Price segmentation'] = wine['Price(Dollars/Unit)'].apply(get_price_seg)

#Добавили новый столбец для ценовой сегментации
```

```python
def get_yiel_cat(yiel):
    if yiel <= 5:
        return 'Poor'
    elif yiel < 15:
        return 'Average'
    elif yiel >= 15:
        return 'High'
    else:
        return yiel

#Запустили функцию для добавления нового столбца
```

```python
wine_production['Production'].max() - wine_production['Production'].min()

#нашли размах переменной Production
```

```python
np.percentile(wine_production['Production'], 75) - np.percentile(wine_production['Production'], 25)

#нашли интерквартильный размах переменной Production
```

```python
wine_mean = wine_production['Production'].mean()
deviations = wine_production['Production'] - wine_mean
sq_deviations = deviations ** 2
variance_samp = sq_deviations.sum() / (len(wine_production) - 1)
std_samp = np.sqrt(variance_samp)
print(variance_samp, std_samp)

#нашли среднеквадратическое отклонение переменной Production
```

```python
variance_pop = sq_deviations.sum() / (len(wine_production))
std_pop = np.sqrt(variance_pop)
print(variance_pop, std_pop)


#нашли дисперсию среднеквадратичного отклонения генеральной совокупности 
```

```python
q3 = np.percentile(wine['Value(Dollars)'], 75)
q1 = np.percentile(wine['Value(Dollars)'], 25)
print(q3, q1)
```

```python
igr = q3 - q1
print(igr)
```

```python
igr_threshold_botton = q1 - 1.5 * igr
igr_threshold_top = q3 + 1.5 * igr
print(igr_threshold_botton, igr_threshold_top)
```
